### Retira os casos normais se necessário

```python
# cria o diretório caso não exista
def createfolder(path):
    if not os.path.exists(path):
        os.makedirs(path)

# se houver a pasta, ela é removida
import shutil
def remove_folders(path):
    if os.path.exists(path):
        shutil.rmtree(path)
```

```python
def remove_cases(df):
    # df2 = df_train.copy()
    indexNames = df[(df['class_name'] == 'negative')].index
    df.drop(indexNames , inplace=True)
    df = df.reset_index(drop=True) # reinicia o index
    return df
```

```python
# caso = 1 => não faz alteração, pois contém imagens de casos normais no treino/val e no teste
# caso = 2 => remove os casos normais do treino/val e mantém no teste
# caso = 3 => remove os casos normais do treino/val e do teste (mantém apenas covid)

caso = 3
if caso == 1:
    path_csv = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/2class_trainCovidNormal/'
elif caso == 2:
    path_csv = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/2class_trainCovid/'
elif caso == 3:
    path_csv = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/1class_trainCovid/'
    

df_train = pd.read_csv(path_csv + 'train.csv')
df_val = pd.read_csv(path_csv + 'val.csv')
df_test = pd.read_csv(path_csv + 'test.csv')

# Remove os casos negativos do conj de treino
if caso == 1:
    print('Size train: ', len(df_train), '\nSize val: ', len(df_val), '\nSize test: ', len(df_test))
if caso == 2:
    df_train = remove_cases(df_train)
    df_val = remove_cases(df_val)
    print('Size train: ', len(df_train), '\nSize val: ', len(df_val), '\nSize test: ', len(df_test))
elif caso == 3:
    df_train = remove_cases(df_train)
    df_val = remove_cases(df_val)
    df_test = remove_cases(df_test)    
    print('Size train: ', len(df_train), '\nSize val: ', len(df_val), '\nSize test: ', len(df_test))
```

```python
from PIL import Image
def copy_files(df, source, dest, ext):
    for i in range(len(df)):
        name = df['id'][i].replace('_image', ext)
        createfolder(dest)
        shutil.copyfile(source + name, dest + name)
#         im = Image.open(path_dataset + name)        
#         im.save(dest + name)
```

```python
path_data = '/home/karem/Artigo_Periodico/Detector/YoloV5/1_Divide_Dados_Train_Test/'

# Imagens
source_images = path_data + 'images_' + str(caso) + '/'
destination_images = '/home/karem/Artigo_Periodico/Detector/YoloV5/sample_data/images/'

remove_folders(destination_images)

copy_files(df_train, source_images + 'train/', destination_images + 'train/', '.png')
copy_files(df_val, source_images + 'valid/', destination_images + 'val/', '.png')
copy_files(df_test, source_images + 'test/', destination_images + 'test/', '.png')


# labels
source_labels = path_data + 'labels_' + str(caso) + '/'
destination_labels = '/home/karem/Artigo_Periodico/Detector/YoloV5/sample_data/labels/'

remove_folders(destination_labels)

copy_files(df_train, source_labels + 'train/', destination_labels + 'train/', '.txt')
copy_files(df_val, source_labels + 'val/', destination_labels + 'val/', '.txt')
copy_files(df_test, source_labels + 'test/', destination_labels + 'test/', '.txt')
```
