```python
import pandas as pd
import os
```

```python
def create_folders(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
# se houver a pasta, ela é removida
import shutil
def remove_folders(path):
    if os.path.exists(path):
        shutil.rmtree(path)
        
# Converte string para lista
def Convert_str_to_int(string):
    li = list(string.split(", "))
    return li

import re
def convert_df_lista_int(str1):
    #str1 = df['xmin_ori'][1]
    str1 = re.sub(r"[\([{})\]]", "", str1) # remove os caracteres da string
    #str1 = (str1.replace('.0', '')) # substirui a casa decimal .0 por vazio, tornando a string no formato int
    str1 = Convert_str_to_int(str1) # converte a string em lista
    #str1 = list(map(float, str1)) # converte a lista de string em float
    #str1 = list(map(int, str1)) # converte a lista de float em int
    str1 = [int(float(i)) for i in str1] # convert os valores para inteiro e depois string    
    return str1
```

```python
# '''
# carrega o arquivo contendo o tamanho original das imagens
# '''

# caso = 1
# caso = 2
caso = 3

# diretorio = os.getcwd() + '/'

if caso == 1:
    diretorio = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/2class_trainCovidNormal/'
elif caso == 2:    
    diretorio = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/2class_trainCovid/'
elif caso == 3:
    diretorio = '/home/karem/Artigo_Periodico/1_Dataset/CSV_Files/1class_trainCovid/'

dir2 = os.getcwd() + '/'
# print(dir2)

folder = ['train', 'val', 'test']

#im_format = '.png'
im_format = '.jpg'
```

```python
for i in range(len(folder)):
    
    ## df = pd.read_csv(diretorio + 'csv_covid_normal/' + folder + '.csv')
    df = pd.read_csv(diretorio + folder[i] + '.csv')
    print(len(df))
    # df
    
    
    dir_ori = dir2 + 'images/' + folder[i] + '/'
    dir_dest = dir2 + '/labels_xml/' + folder[i] + '/'
    
    remove_folders(dir_dest)
    create_folders(dir_dest)
    
    
    for i in range(len(df)):
    
        name = df['id'][i].replace('_image', '')    
        path_image = dir_ori + name + im_format

        width = df['width'][i] 
        height = df['height'][i] 

        filename = name + im_format

        #print(i , name)

        # write in xml file
        with open(dir_dest + name + '.xml' , "w") as xml_file: 
            xml_file.write('<annotation>\n')
            xml_file.write('    <folder></folder>\n')
            xml_file.write('    <filename>' + filename + '</filename>\n')
            #xml_file.write('    <path>' + path_image + '</path>\n')
            xml_file.write('    <path>' + filename + '</path>\n')
            xml_file.write('    <size>\n')
            xml_file.write('        <width>' + str(width) + '</width>\n')
            xml_file.write('        <height>' + str(height) + '</height>\n')
            xml_file.write('        <depth>3</depth>\n')
            xml_file.write('    </size>\n')
            xml_file.write('    <segmented>'+str(0)+'</segmented>\n')        


            lista_int_x1 = convert_df_lista_int(df['xmin'][i])
            lista_int_x2 = convert_df_lista_int(df['xmax'][i])
            lista_int_y1 = convert_df_lista_int(df['ymin'][i])
            lista_int_y2 = convert_df_lista_int(df['ymax'][i])


            for j in range(len(lista_int_x1)):

                if df['class_name'][i] == 'indeterminate' or df['class_name'][i] == 'typical':
                    classe_xml = 'covid'
                else:
                    classe_xml = 'normal'
                    #print('entrou')

                xml_file.write('    <object>\n')
                xml_file.write('        <name>' + classe_xml + '</name>\n')
                xml_file.write('        <pose>Unspecified</pose>\n')
                xml_file.write('        <truncated>0</truncated>\n')
                xml_file.write('        <difficult>0</difficult>\n')
                xml_file.write('        <occluded>0</occluded>\n')
                xml_file.write('        <bndbox>\n')
                if(lista_int_x1[j]) < 0:
                    lista_int_x1[j] = 0                
                xml_file.write('            <xmin>' + str(lista_int_x1[j]) + '</xmin>\n')
                xml_file.write('            <xmax>' + str(lista_int_x2[j]) + '</xmax>\n')
                xml_file.write('            <ymin>' + str(lista_int_y1[j]) + '</ymin>\n')
                xml_file.write('            <ymax>' + str(lista_int_y2[j]) + '</ymax>\n')
                xml_file.write('        </bndbox>\n')
                xml_file.write('    </object>\n')

            xml_file.write('</annotation>')
```

```python
# renomeia a pasta val (se ela existir) -> arquivo original é valid
if os.path.exists('labels_xml/val'):
    os.rename('labels_xml/val', 'labels_xml/valid')
```
