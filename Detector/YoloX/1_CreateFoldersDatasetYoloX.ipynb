```python
# se houver a pasta, ela é removida
import shutil
import os
def remove_folders(path):
    if os.path.exists(path):
        shutil.rmtree(path)
        
# cria o diretório caso não exista
def createfolder(path):
    if not os.path.exists(path):
        os.makedirs(path)

        
# Captura o diretório atual e salva na variável
path = diretorio = os.getcwd() + '/'
print(path)

folder = 'train/'

# Remove pastas remanescentes de treinamento antigo da YoloX
remove_folders(path + folder) # remove pastas/diretórios
# if os.path.exists(path + 'name file.ext'): # remove arquivo
#     os.remove(path + 'name file.ext')
```

```python
import subprocess

def copia(source_path, dest_data):
    ## write an rsync commands to merge the directories
    rsync_cmd = 'rsync' + ' -avzh ' + source_path + ' ' + dest_data

    ## run the rsync command
    subprocess.run(rsync_cmd, shell=True)
```

```python
##############################################
### Criar pasta VOC2007
##############################################

### Folder: "Annotations" 
createfolder(path + folder + '/VOC2007/Annotations/')

# Copiar conteúdo da pasta /Artigo_Periodico/YoloX/Dados/labels_xml
# Juntar tudo (train, val e test) na pasta "Annotations"

## Define a pasta
conj = ['train/', 'valid/', 'test/']

for i in range (len(conj)):
    ## define your paths
    source_path = path + '1_Dados/labels_xml/' + conj[i]       
    dest_data = path + folder + '/VOC2007/Annotations/' ## where to place the merged data    
    
    copia(source_path, dest_data)       
```

```python
### Folder: "JPEGImages"
createfolder(path + folder + '/VOC2007/JPEGImages/')

# Copiar conteúdo da pasta /Artigo_Periodico/YoloX/Dados/images
# Juntar tudo (train, val e test) na pasta "JPEGImages"

## Define a pasta
#conj = ['train/', 'valid/', 'test/']

for i in range (len(conj)):
    ## define your paths
    source_path = path + '1_Dados/images/' + conj[i]       
    dest_data = path + folder + '/VOC2007/JPEGImages/' ## where to place the merged data    
    copia(source_path, dest_data)   
```
